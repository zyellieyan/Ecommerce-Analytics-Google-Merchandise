### kfold模型

```python
kf = KFold(n_splits=5)
KFold(n_splits=5, random_state=1, shuffle=False)
rmse_all = np.zeros(5)
i=0
for train_index, test_index in kf.split(X):
    #建模
    rf = RandomForestRegressor(n_estimators=1000,random_state=0)
    #训练
    rf.fit(X[train_index],y[train_index])
    #prediction
    prediction_test = rf.predict(X[test_index])
    prediction_train = rf.predict(X[train_index])
    #rmse
    rmse_test = sqrt(mean_squared_error(y[test_index], prediction_test))
    rmse_train = sqrt(mean_squared_error(y[train_index], prediction_train))
    print('-'*20,i,'-'*20)
    print('rmse - test: ',rmse_test, ' '*10, 'rmse - train: ',rmse_train)
    rmse_all[i]=rmse_test
    i += 1 
print('rmse_mean: ', np.mean(rmse_all))
```

```python
from sklearn.model_selection import RandomizedSearchCV
# Number of trees in random forest
n_estimators = [300,2000]
# Number of features to consider at every split
max_features = ['auto', 'sqrt',7]
# Maximum number of levels in tree
max_depth = [17,18,19,20]
# Minimum number of samples required to split a node
min_samples_split = [33,38,41]
# Minimum number of samples required at each leaf node
min_samples_leaf = [16,17,18,29]
# Method of selecting samples for training each tree
bootstrap = [True, False]
# Create the random grid
random_grid = {'n_estimators': n_estimators,
               'max_features': max_features,
               'max_depth': max_depth,
               'min_samples_split': min_samples_split,
               'min_samples_leaf': min_samples_leaf,
               'bootstrap': bootstrap}
pprint(random_grid)
```

```python
# Use the random grid to search for best hyperparameters
# First create the base model to tune
rf = RandomForestRegressor(random_state = 20)
# Random search of parameters, using 5 fold cross validation, 
# search across 100 different combinations, and use all available cores
rf_random = RandomizedSearchCV(estimator=rf, param_distributions=random_grid,
                              n_iter = 10, scoring='neg_root_mean_squared_error', 
                              cv = 5, verbose=3, random_state=20, n_jobs=-1,
                              return_train_score=True)

# Fit the random search model
rf_random.fit(X, y);
```
