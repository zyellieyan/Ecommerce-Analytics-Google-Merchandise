```python
import pandas as pd 
import numpy as np
pd.set_option('max_columns',100)
pd.set_option('max_rows',100)

import matplotlib.pyplot as plt
%matplotlib inline
```

```python
data1_train = pd.read_csv('data1_train.csv')
data1_train.head()
```

### for train

```python
data1_train = data1_train.drop(labels =['weenkend','weekday','Unnamed: 0'],axis=1)
data1_train.head()
```

### for test

```python
data1_test = pd.read_csv('data1_test.csv')
data1_test.head()
```

```python
data1_test = data1_test.drop(labels = ['weenkend','weekday','Unnamed: 0'],axis=1)
data1_test.head()
```

```python
print(data1_test.shape)
```

#### 分出train的X，y

```python
X_train = data1_train.drop(labels=['fullVisitorId','logTransaction'],axis=1)

feature_list = list(X_train.columns)

X_train = X_train.values
y_train=data1_train['logTransaction'].values 
print(X_train.shape,y_train.shape)
```

#### 分出test的X，y

```python
X_test = data1_test.drop(labels=['fullVisitorId','logTransaction'],axis=1)
X_test = X_test.values

y_test=data1_test['logTransaction'].values 
y_test.shape

print(X_test.shape,y_test.shape)
```

### 建模

```python
from sklearn.ensemble import RandomForestRegressor
#记住这里是在做regression，所以import的是randomforestregressor
```

```python
#建模
rf = RandomForestRegressor(n_estimators=1000,random_state=0)

#训练
rf.fit(X_train,y_train)
```

### 检测 MAE

```python
from sklearn.metrics import mean_absolute_error
```

```python
# predict the result for test
prediction_test = rf.predict(X_test)

mae_test=mean_absolute_error(y_test,prediction_test)

print('MAE',mae_test)
```

```python
# predict the result for train
prediction_train = rf.predict(X_train)

mae_train=mean_absolute_error(y_train,prediction_train)

print('MAE',mae_train)
```

### RMSE

```python
from sklearn.metrics import mean_squared_error
from math import sqrt

rmse_test = sqrt(mean_squared_error(y_test, prediction_test))
rmse_train = sqrt(mean_squared_error(y_train, prediction_train))

print('rmse - test: ',rmse_test,"    rmse - train: ",rmse_train)
```

### feature importance

```python
# get the feature importance
importances = list(rf.feature_importances_)

# transform the formula
feature_importances = [(feature, importance) for feature, importance in zip(feature_list, importances)]

# descending order
feature_importances = sorted(feature_importances, key= lambda x:x[1], reverse=True)

# print out
for a in feature_importances:
    print('Variable: {}      Importance:{}'.format(a[0],a[1]))
```

```python
# 转换成list格式
x_values = range(len(importances))

# 绘图
plt.bar(x_values, importances, orientation = 'vertical',facecolor = 'mediumpurple',alpha = 0.6,width=0.8)

# x轴名字
plt.xticks(x_values, feature_list, rotation='vertical')

# 图名
plt.ylabel('Importance'); plt.xlabel('Feature'); plt.title('Feature Importances'); 
```

### time trade-offs

```python
# 要计算时间了
import time

# 这次是用所有特征
all_features_time = []


# 算一次可能不太准，来5次取个平均
for i in range(5):
    start_time = time.time()
    rf.fit(X_train,y_train)
    all_features_predictions = rf.predict(X_test)
    end_time = time.time()
    all_features_time.append(end_time - start_time)

all_features_time = np.mean(all_features_time)
print('使用所有特征时建模与测试的平均时间消耗:', round(all_features_time, 2), '秒.')
```

```python
# 这次是用部分重要的特征
reduced_features_time = []
rf = RandomForestRegressor(n_estimators=1000,random_state=0)
# 算一次可能不太准，来5次取个平均
for i in range(5):
    start_time = time.time()
    rf_imp.fit(X_train_imp,y_train_imp)
    reduced_features_predictions = rf_imp.predict(X_test_imp)
    end_time = time.time()
    reduced_features_time.append(end_time - start_time)

reduced_features_time = np.mean(reduced_features_time)
print('使用所有特征时建模与测试的平均时间消耗:', round(reduced_features_time, 2), '秒.')
```

```python
# 用分别的预测值来计算评估结果
all_rmse =  sqrt(mean_squared_error(y_final_test, prediction_final_test))
reduced_rmse = sqrt(mean_squared_error(y_true_test_imp,prediction_final_test_imp))

#创建一个df来保存结果
comparison = pd.DataFrame({'features': ['all (22)', 'reduced (15)'], 
                           'run_time': [round(all_features_time, 3), round(reduced_features_time, 3)],
                           'MAE':[round(mae_final_test, 3), round(mae_final_test_imp, 3)],
                           'RMSE': [round(all_rmse, 3), round(reduced_rmse, 3)]})

comparison[['features', 'RMSE', 'MAE','run_time']]
```

```python
relative_RMSE_increase = 100 * (reduced_rmse - all_rmse) / all_rmse
print('relative_RMSE_increase:', round(relative_RMSE_increase, 3), '%.')

relative_runtime_decrease = 100 * (all_features_time - reduced_features_time) / all_features_time
print('relative_runtime_decrease:', round(relative_runtime_decrease, 3), '%.')
```
